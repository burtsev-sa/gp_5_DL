```python
import pandas as pd
import os
```

Получим имена всех собранных датасетов

```python
# https://sky.pro/wiki/media/rabota-s-fajlami-v-python-kak-poluchit-spisok-vseh-fajlov-v-direktorii/
kolesa_datasets = os.listdir('data/raw')
```

```python
df_parsed = None
for filename in kolesa_datasets:
    if df_parsed is None:
        df_parsed = pd.read_csv('data/raw/' + filename)
        continue

    df_temp = pd.read_csv('data/raw/' + filename)
    df_parsed = pd.concat([df_parsed, df_temp], axis=0)

```

```python
df_parsed
```

```python
df_parsed = df_parsed.drop_duplicates('kolesa_id')
```

```python
df_parsed.to_csv('fourth_df.csv', index=False)
```
