## Описание признаков

- kolesa_id - уникальный id объявления со стороны сервиса
- kolesa_url - ссылка на объявление
- parsed_at - дата и время получения объявления
- brand - марка авто
- model - модель авто
- generation - поколение данного авто (этой марки)
- year - год выпуска авто
- city - город публикации объявление
- body_type - тип кузова авто
- fuel_type - тип топлива
- engine_volume - объем двигателя
- mileage - пробег в км
- transmission - тип коробки/трансмиссии
- drive_type - тип привода
- steering_wheel - расположение руля
- color - цвет авто    
- kz_registration - растаможен ли в Казахстане
- imgs_count - количество фотографий в объявлении
- price - цена продажи
- img_filename - название файла с главным фото объявления
- img_url - ссылка на на главное фото объявления
- found_img - нашли ли фото

Из них технические: kolesa_id, kolesa_url, parsed_at, img_filename, img_url, found_img

Все остальные могут быть рассмотрены для решения задачи DL на табличных данных

## Подгтовим данные для дальнейшей работы

Общая проверка на пропуски

```python
df.isna().sum()
```

Основная проблема - пропуски в color. Это было обнаружено еще в ходе парсинга. Заполним их значением 

Пропуски в img_filename и img_url мы будем отдельно обрабатывать позже

```python
df['color'] = df['color'].fillna('unknown')
```

### City, body_type, transmission, drive_type, steering_wheel, color, kz_registration

```python
df[['city', 'body_type', 'transmission', 'drive_type', 'steering_wheel', 'color', 'kz_registration']]
```

Все это признаки, содержащие текст

Заметим, что у нас не убраны символы переноса строк \n и, возможно, лишние пробелы

Уберем это у всех этих признаков

```python
df['color'].value_counts()
```

```python
df['city'] = df['city'].apply(lambda x: x.replace('\n', '').strip())
df['body_type'] = df['body_type'].apply(lambda x: x.replace('\n', '').strip())
df['transmission'] = df['transmission'].apply(lambda x: x.replace('\n', '').strip())
df['drive_type'] = df['drive_type'].apply(lambda x: x.replace('\n', '').strip())
df['steering_wheel'] = df['steering_wheel'].apply(lambda x: x.replace('\n', '').strip())
df['color'] = df['color'].apply(lambda x: x.replace('\n', '').strip())
df['kz_registration'] = df['kz_registration'].apply(lambda x: x.replace('\n', '').strip())
```
